# Pillar 3: C3 Pedagogical Mastery Audit (BKT)
**Objective**: Demonstrate how the system tracks the child's actual learning progress over time using Bayesian Knowledge Tracing (BKT).

### Parameters:
- **P(Lo)**: Initial probability of knowing the letter.
- **P(T)**: Probability of learning it during a task.
- **P(S)**: Probability of 'Slipping' (knowing it but making an error).
- **P(G)**: Probability of 'Guessing' (not knowing it but getting it right).

```python
import numpy as np
import matplotlib.pyplot as plt

def update_bkt(p_known, is_correct, p_transit=0.1, p_slip=0.1, p_guess=0.2):
    if is_correct:
        p_obs = (p_known * (1 - p_slip)) / (p_known * (1 - p_slip) + (1 - p_known) * p_guess)
    else:
        p_obs = (p_known * p_slip) / (p_known * p_slip + (1 - p_known) * (1 - p_guess))
    
    new_p_known = p_obs + (1 - p_obs) * p_transit
    return new_p_known

# Simulate learning progress over 15 attempts
history_correct = [0, 0, 1, 0, 1, 1, 0, 1, 1, 1, 1, 1, 1, 1, 1]
p_values = [0.15]

for correct in history_correct:
    p_values.append(update_bkt(p_values[-1], correct))

plt.figure(figsize=(10, 5))
plt.plot(p_values, 'o-', color='green', label='Mastery Probability P(K)')
plt.axhline(y=0.95, color='red', linestyle='--', label='Mastery Threshold')
plt.title('C3 Engine: Bayesian Knowledge Tracing (BKT) Progress', fontsize=13)
plt.xlabel('Number of Interaction Attempts', fontsize=11)
plt.ylabel('Probability of Knowledge Mastery', fontsize=11)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
